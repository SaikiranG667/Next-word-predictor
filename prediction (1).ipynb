```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense,Masking
```

```python
with open("/content/drive/MyDrive/Next_word_prediction/data/text.txt",'r',encoding='utf-8') as file:
  text=file.read()
paragraphs = text.split('\n\n')

paragraphs = [para.strip() for para in paragraphs if para.strip()]
```

```python
paragraphs = [sentence.replace('\n', ' ') for sentence in paragraphs]
```

```python
paragraphs=np.array(paragraphs)
```

```python
paragraphs
```

```python
tokenizer=Tokenizer()

new=tokenizer.fit_on_texts(
    paragraphs
)
```

```python

print(tokenizer.word_index)
```

```python
sequences = tokenizer.texts_to_sequences(paragraphs)
```

```python
last_index = max(tokenizer.word_index.values())
last_index
```

```python
len(sequences)
```

```python
sequences
```

```python
ip_sequences=[]
for i in sequences:
      for j in range(1,len(i)):
            ip_sequences.append(i[:j+1])
```

```python
ip_sequences
```

```python
max_len=max([len(x) for x in ip_sequences])
max_len
```

```python
X = [lst[:-1] for lst in ip_sequences]
Y = [lst[-1] for lst in ip_sequences]
```

```python
len(X)
```

```python
len(Y)
```

```python
max_len=max([len(x) for x in X])
max_len
```

```python
max(Y)
```

```python
Y=np.array(Y)
```

```python
from tensorflow.keras.utils import to_categorical
Y=to_categorical(Y,num_classes=8923)
```

```python
model = Sequential()
model.add(Masking(mask_value=0.0, input_shape=(465,)))
model.add(Embedding(input_dim=8922, output_dim=100))
model.add(LSTM(500,return_sequences=True))
model.add(LSTM(300))
model.add(Dense(256,activation='relu'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X,Y,epochs=100,validation_split=0.2,batch_size=64)
```
